- Q1: File sentences.csv contains 15 sentences, highlighting different mood of an individual. These moods are classified into two categories depressed and casual, based on the words listed below, depressed=[rejects, bullied, lonely, dead, wrong] casual = [excited, happy, party, good] Use this list to add label column as Condition either depressed or normal. 

```python
depressed = ["rejects", "bullied", "lonely", "dead", "wrong"] 
casual = ["excited", "happy", "party", "good"]
```

```python
mood = []
for i in df["Sentences"]:
    for j in casual:
        if j in i:
            mood.append("casual")
            break
    else: #else for for loop, when break is used
        for k in depressed:
            if k in i:
                mood.append("depressed")
                break
        else:
            mood.append("NONE")
```

```python
df["mood"]=mood
```
